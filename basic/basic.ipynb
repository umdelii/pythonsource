```python
# 동전 교환하기
# 7,777원 => 500원, 100원, 50원, 10원으로 몇 개?
x = 7777
a = x//500
b = (x-(500*a))//100
# b = (x%500)//100
c = (x-(500*a)-(b*100))//50
d = (x-(500*a)-(b*100)-(c*50))//10
print(a,b,c,d)
print('500원 %d개, 100원 %d개, 50원 %d개, 10원 %d개' % (a,b,c,d))
print(f'500원 {a}개, 100원 {b}개, 50원 {c}개, 10원 {d}개')
```

```python
# 다중조건
# if else if else => if elif else
num = 90

if num >= 90:
    print('A')
elif num >= 80:
    print('B')
elif num >= 70:
    print('C')
elif num >= 60:
    print('D')
else: print('E')
```

```python
# 사칙연산(숫자 2개, 연산자 입력받기)
num1 = int(input('num1 :'))
num2 = int(input('num2 :'))
cal = input('연산자 입력 :')

if cal == '+':
    print(num1+num2)
elif cal == '-':
    print(num1-num2)
elif cal == '*':
    print(num1*num2)
elif cal == '/':
    print(num1/num2)
elif cal == '//':
    print(num1//num2)
elif cal == '%':
    print(num1%num2)
```
