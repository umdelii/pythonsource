```python
class Car:
    color = ""
    speed = 0

    def upSpeed(self, value):
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

    def __str__(self):  # == toString()
        return self.color + " , " + str(self.speed)
```

```python
car1 = Car()
car1.color = "red"
car1.speed = 0

car2 = Car()
car2.color = "black"
car2.speed = 0

car1.upSpeed(90)
car2.upSpeed(110)

print(car1)
print(car2)
```

```python
# 클래스 메소드
class SelfTest:
    @classmethod
    def func1(cls):  # 클래스 메소드 : self 매개변수로 선언x
        print("function1 called")

    def func2(self):  # 인스턴스 메소드
        print("function2 called")
```

```python
obj1 = SelfTest()
# obj1.func1() TypeError: SelfTest.func1() takes 0 positional arguments but 1 was given
SelfTest.func1()
obj1.func2()
```

```python
class SelfTest2:
    num = 15

    @staticmethod  # 스태틱 메소드 : 내부 변수(클래스변수,self...) 쓰지 않아도 됨 / 외부 변수 사용
    def func1(a, b):
        return a + b

    def func2(self):
        print("func2 called")

    @classmethod
    def func3(cls):
        print(f"class method {cls.num}")
```

```python
test = SelfTest2()
test.func2()

SelfTest2.func3()

print(SelfTest2.func1(1, 2))
test.func1(1, 2)
```

```python
class UserInfo2:

    # 생성자 오버로딩 없음
    # default 값을 사용해서 오버로딩과 유사한 효과
    def __init__(self, name, age, email=None):
        self.name = name
        self.age = age
        self.email = email

    def user_info(self):
        print("메소드 실행")

    def __str__(self):
        return f"name : {self.name}, age : {self.age}, email : {self.email}"
```

```python
# email이 필수 입력 요소가 아닐 때
user1 = UserInfo2("정이안", 18)
user2 = UserInfo2("최지우", 20, "applepie@google.com")

print(user1.__dict__)
print(user2)
```

```python
class Car:
    def __init__(self, color="black", speed=0):
        self.color = color
        self.speed = speed

    def upSpeed(self, value):
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

    def __str__(self):  # == toString()
        return self.color + " , " + str(self.speed)
```

```python
class Parent:
    def __init__(self):
        self.value = "test"

    def test(self):
        print("Parent test() 호출")


class Child(Parent):
    def __init__(self):
        # Parent.__init__(self) # 부모의 인스턴스 변수 호출여부
        super().__init__()
        print("Child 클래스의 __init__ 호출")
```

```python
parent = Parent()
parent.test()

child = Child()
child.test()
print(child.value)
```

```python
class AudioVisual:
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume

    def switch(self, onOff):
        self.power = onOff

    def set_volume(self, volume):
        self.volume = volume


class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        str1 = "La la la..." if self.power else "turn it on"
        print(str1)


class TV(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        str1 = "have fun!" if self.power else "switch on"
        print(str1)
```

```python
obj1 = TV(False, 10, 40)
obj1.switch(True)
obj1.watch()

obj2 = Audio(False, 8)
obj2.set_volume(10)
obj2.tune()
```

```python
class Parent:
    def __init__(self):
        self.value = "test"

    def func1(self):
        print("Parent func1() 호출")

    def func2(self):
        print("Parent func2() 호출")


class Child(Parent):
    def __init__(self):
        # Parent.__init__(self) # 부모의 인스턴스 변수 호출여부
        super().__init__()
        print("Child 클래스의 __init__ 호출")

    def func1(self):
        print("Parent func1() 재정의")

    def func3(self):
        print("Child func3() 호출")
```

```python
parent, child = Parent(), Child()
parent.func1()
parent.func2()

child.func1()
child.func2()
child.func3()
```

```python
# private 키워드는 없음
# __변수 : private
import math


class Circle:
    def __init__(self, radius):
        self.__radius = radius

    def get_circ(self):
        return 2 * math.pi * self.__radius

    def get_area(self):
        return math.pi * (self.__radius**2)
    
    def get_radius(self):
        return self.__radius
```

```python
circle = Circle(10)
print(f"둘레 {circle.get_circ():.4f}")
print(f"면적 {circle.get_area():.4f}")
# print(circle.__radius) AttributeError: 'Circle' object has no attribute '__radius'
print(circle.get_radius())
```
