- word.txt 읽은 후 리스트 생성
- 리스트 밸류 무작위 섞은 후 랜덤으로 단어 하나 추출, 출력
- 사용자는 보여진 단어 입력
- 총 5번 반복
- 정확히 타이밍한 단어가 3자 이상 ? '합격' or '불합격
- 게임시간과 정답개수 출력

```python
import time

start = time.time()
time.sleep(3)
end = time.time()

print(end - start)
```

```python
import math, random, time

fr = open("./data/word.txt", "r", encoding="utf-8")
word_list = [w.strip() for w in fr.readlines()]
random.shuffle(word_list)

input("Game Start")
start = time.time()
cor_count = 0
for i in range(5):
    word = random.choice(word_list)
    print(word)

    input_user = input("다음 제시된 단어를 똑같이 입력하시오")
    if word == input_user:
        cor_count += 1
        print("----correct----")

end = time.time()
print(f"게임시간 : {end-start:.2f}, 정답개수 : {cor_count}")

if cor_count > 3:
    print("합격")
else:
    print("불합격")

fr.close()
```
