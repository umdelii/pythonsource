```python
# 주차장 프로그램
# 5대만 주차 가능 (차례대로 입차)

parking_lot = []
top, car_name = 0, "A"

while True:
    no = int(input("[1] 자동차 넣기 | [2] 자동차 빼기 | [3] 프로그램 종료"))

    if no <= 3:
        if no == 1:
            if top >= 5:
                print("만차")
            else:
                parking_lot.append(car_name)
                car_name = chr(ord(car_name) + 1)
                top += 1
                print(f"주차장 상태 {parking_lot}")
        elif no == 2:
            if top > 0:
                parking_lot.pop()
                car_name = chr(ord(car_name) - 1)
                top -= 1
                print(f"주차장 상태 {parking_lot}")
            else:
                print("출차 할 차 없음")
        elif no == 3:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인하십시오")
```

### set : 집합
- 중복 불가 
- 순서가 없음 : 인덱싱 불가

```python
s1 = set()
s2 = set([1,2,3])
s3 = set([1,2,3,3,3])
s4 = set('Hello')
```

```python
print(s1)
print(s2)
print(s3)
print(s4)
```

```python
# 인덱싱을 하고싶다면? 형변환
t1 = tuple(s2)
t1
s2

list1 = list(s3)
list1[2]
```

```python
# 교집합, 합집합, 차집합
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1 & s2)
print(s1.intersection(s2))

print(s1 | s2)
print(s1.union(s2))

print(s1 - s2)
print(s1.difference(s2))

#대칭 차집합
print(s1 ^ s2)
print(s1.symmetric_difference(s2))
```
