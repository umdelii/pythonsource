```python
url = "https://google.com/"

# urlretrieve : 요청하는 url의 정보를 로컬파일로 저장, 반환 타입 : 튜플
# csv, api 데이터 등 많은 양의 데이터를 한 번에 저장할 때 사용
file1, header1 = req.urlretrieve(url, "./data/google.html")
print(header1)
```

```python
# urlopen() : 다운로드 x, 메모리에 올려놓고 분석
url = "https://naver.com"
response = req.urlopen(url)

print(f"url 정보 {response.geturl()}")
print(f"url 상태 {response.status}")
print(f"url headers 정보 {response.getheaders()}")

# 내용(content) 가져오기
# contents = response.read()
contents1 = response.read().decode("utf-8")
# print(contents[:4000])
print(contents1[:4000])
```

```python
# ctxCd 파라메터 값 변경
url = "https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp"

userAgent = UserAgent()
fake_headers = {"user-agent": userAgent.random}

# params = {"ctxCd": "1001"}
params = [
    {"ctxCd": "1001"},
    {"ctxCd": "1012"},
    {"ctxCd": "1013"},
]

with requests.Session() as s:

    for p in params:
        res = s.get(url=url, params=p, headers=fake_headers)
        print(res.text)
```

#### BeautifulSoup

- requests(가져오기) + BeautifulSoup(파싱) 조합으로 사용


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://n.news.naver.com/mnews/article/629/0000472448"

res = requests.get(url)
# print(res.text)
soup = BeautifulSoup(res.text, "html.parser")

# 원하는 요소 추출
# 1. 태그명 사용
# print(soup.body)
print(soup.h2)  # 제일 처음 태그
print(soup.h2.text)
print(soup.h2.attrs)
```

```python
with open("./data/story.html", "r") as fr:
    story = fr.read()

soup = BeautifulSoup(story, "html.parser")

# 태그
print(soup.title)
print(soup.title.text)
print(soup.p)
print(soup.p.text)
```

```python
# 문서 구조
title = soup.title
print(title.parent)

children = soup.body.children
print(children)
print([child for child in children])
```

```python
# find~~
# find('기준','속성')
print(f"title {soup.find('title')}")
print(f"p {soup.find('p')}")
print(f"p {soup.find('p',class_='story')}")

print(f"{soup.find('a',id='link1')}")
print(f"{soup.find('p',attrs={'class':'story','data-num':'3'})}")

print(f"{soup.find_all('p',class_='story')}")  # list 형태로 반환
```

```python
# 구조로 찾기
p1 = soup.p

print(p1.find_next_sibling())
```

```python
# select() / select_one() : css selector 사용
title = soup.select_one("p.title")
print(title)

a = soup.select_one("a#link1")
print(a)
print(a.text.strip())
```

```python
tags = soup.select(".sister")
print([tag for tag in tags])
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://pythonscraping.com/pages/page3.html"
soup = BeautifulSoup(requests.get(url=url).text, "html.parser")

# Totally Normal Gifts 출력
print(soup.find("h1").get_text())

# 첫번째 문단 출력
paras = soup.select_one("div#content")
paras.find("p").decompose()
para = paras.text
print(para)

# 테이블 th 출력
ths = soup.select("table#giftList th")
print([th.text.strip() for th in ths])

# cost 출력
costs = soup.select("table#giftList tr.gift td:nth-child(3)")
print([cost.get_text().strip() for cost in costs])

# 이미지 찾기
imgs = soup.find_all("img")
for img in imgs:
    print(img.get("src"))
```
