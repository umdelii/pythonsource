### 동적 크롤링
- 사람이 브라우저를 통해서 시도하는 것처럼 크롤링 가능

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.request as req
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.EdgeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)
driver.maximize_window()

driver.get("http://google.com")

print(f"title: {driver.title}")
print(f"url: {driver.current_url}")
print(f"source: {driver.page_source}")

time.sleep(3)
driver.close()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

driver.get("http://naver.com")

time.sleep(3)

# 검색창에 검색어 입력
# 1. 검색창 찾기
element = driver.find_element(by=By.ID, value="query")

# 2. 검색어 입력
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

# 3. 결과페이지로 이동할 때까지 시간 부여
time.sleep(2)


driver.close()
```

```python
driver = webdriver.Edge()
driver.maximize_window()

driver.get("https://www.python.org/")

time.sleep(3)

element = driver.find_element(By.ID, "id-search-field")
element.send_keys("python")
element.send_keys(Keys.ENTER)

time.sleep(2)

# 4. 파싱 : selenium, BeautifulSoup
titles = driver.find_elements(By.TAG_NAME, "h3")
for title in titles:
    print(title.text)

driver.close()
```

```python
# iframe으로 구성된 페이지 접근

driver = webdriver.Edge()
driver.minimize_window()

driver.get("https://www.w3schools.com/html/tryit.asp?filename=tryhtml_basic_document")

time.sleep(2)

# iframe 안으로 이동
driver.switch_to.frame("iframeResult")

# iframe 안 요소 찾기
element = driver.find_element(By.TAG_NAME, "h1")
print(element.text)

# 다시 iframe밖으로 나오기
driver.switch_to.default_content()

# 왼쪽 요소 찾기
pre = driver.find_element(
    By.CSS_SELECTOR,
    "#textareawrapper > div > div.CodeMirror-scroll > div.CodeMirror-sizer > div > div > div > div.CodeMirror-code > pre:nth-child(7)",
)
print(pre.text)

time.sleep(2)

titles = driver.find_elements(By.TAG_NAME, "h3")
for title in titles:
    print(title.text)

driver.close()
```

```python
# naver.com
# 새탭 daum.net

driver = webdriver.Edge()
driver.maximize_window()

driver.get("http://naver.com")
time.sleep(2)

# 새 탭 띄우기
# window.open('주소') <= js 코드
driver.execute_script("window.open('http://daum.net')")

# 현재 열려진 모든 탭 가져오기
tabs = driver.window_handles
# tabs[0] => naver, [1] => daum

# 다음에서 검색어 입력
driver.switch_to.window(tabs[1])
element = driver.find_element(By.ID, "q")
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

time.sleep(2)

# 네이버에서 검색어 입력
driver.switch_to.window(tabs[0])
element = driver.find_element(by=By.ID, value="query")
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

time.sleep(2)
driver.close()
```

```python
# 스크롤

driver = webdriver.Edge()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(2)

# element = driver.find_element(By.ID, "AKCSearch")
# element.send_keys("흑염소 진액")
# element.send_keys(Keys.ENTER)
# time.sleep(3)

# 스크롤 이동
prev_height = driver.execute_script("return document.body.scrollHeight") # 현재 문서 높이
while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break
    prev_height = curr_height

# 스크롤 이동 완료 후 처음으로 스크롤 이동
driver.execute_script("window.scrollTo(0,0)")

# 파싱
# driver.page_source

time.sleep(2)
driver.close()
```

```python
# 이미지 다운로드
driver = webdriver.Edge()
driver.get("http://naver.com")

time.sleep(3)

element = driver.find_element(By.ID, "query")
element.send_keys("바이올린")
element.send_keys(Keys.ENTER)

time.sleep(2)

# 이미지 탭 클릭
driver.find_element(
    By.XPATH, '//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[1]/a'
).click()
time.sleep(2)

# 스크롤 이동
prev_height = driver.execute_script(
    "return document.body.scrollHeight"
)  # 현재 문서 높이
while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break
    prev_height = curr_height

# 스크롤 이동 완료 후 처음으로 스크롤 이동
driver.execute_script("window.scrollTo(0,0)")

time.sleep(3)

# 이미지 저장
images = driver.find_elements(
    By.CSS_SELECTOR, ".thumb .link_thumb"
)  # 전체 썸네일 이미지

count = 1
for img in images:
    img.click()
    time.sleep(2)

    # 큰 이미지 저장
    large_image = driver.find_element(
        By.CSS_SELECTOR, "div.viewer_image img"
    ).get_attribute("src")
    req.urlretrieve(large_image, f"./download/violin{count}.png")

    count += 1


driver.close()
```

```python
url = "https://prod.danawa.com/list/?cate=112758&15main_11_02="

driver = webdriver.Edge()
driver.get(url=url)
time.sleep(2)

# 광고창 닫기
driver.find_element(
    By.XPATH, '//*[@id="newProductBanner"]/div/div/div[3]/button[1]/span'
).click()
time.sleep(2)

# M4맥북 클릭
driver.find_element(By.XPATH, '//*[@id="viewMore_43778"]/span').click()
time.sleep(2)

# 제품명, 이미지, 가격
cur_page = 1
target_crawl_num = 0
while True:
    products = driver.find_elements(
        By.CSS_SELECTOR, "ul.product_list li.prod_item .prod_main_info"
    )

    print(f"현재 페이지 {cur_page}")

    for product in products:
        name = product.find_element(By.CSS_SELECTOR, "p.prod_name a").text

        # img = product.find_element(By.CSS_SELECTOR, ".thumb_image img").get_attribute(
        #     "src"
        # )
        img = product.find_element(By.CSS_SELECTOR, ".thumb_image img")

        if img.get_attribute("src"):
            src = img.get_attribute("src")
        else:
            src = img.get_attribute("data-original")

        if "https:" not in src:
            src = "https:" + src

        try:
            price = product.find_element(
                By.CSS_SELECTOR, ".prod_pricelist li.rank_one strong"
            ).text
        except NoSuchElementException:
            price = 0

        print(f"{name}, {price}, {img}")

    cur_page += 1

    if cur_page > target_crawl_num:
        break

    # 페이지 번호 클릭
    driver.find_element(
        By.CSS_SELECTOR, f"div.number_wrap a:nth-child{cur_page}"
    ).click()

    time.sleep(2)

driver.close()
```
