# 정규표현식
- 정규식 패턴은 동일

```python
data = """ \
park 800905-1184227
kim 700901-2042397
"""

# 뒷자리를 *******로 출력

# \n을 기준으로 문자열 분리
infos = data.split("\n")
print(infos)

result = []

# 공백을 기분으로 다시 분리
for line in infos:
    word_result = []
    # info = line.strip().split(" ")
    # print(info)

    for info in line.strip().split(" "):
        if (
            len(info) == 14 and info[:6].isdigit() and info[7:].isdigit()
        ):  # isdigit() : 숫자맞음?
            info = info[:6] + "-" + "*******"
        word_result.append(info)

    result.append(" ".join(word_result))

print("\n".join(result))
```
