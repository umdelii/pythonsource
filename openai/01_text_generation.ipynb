```python
# í† í° í™•ì¸
# input_tokens=66
# output_tokens=3880
# total_tokens=3946

response.usage
```

```python
# gpt-5-nano : prices per 1M tokens
# input : $0.05, output : $0.40, cached input : $0.005

total_bill = (
    response.usage.input_tokens / 1000000 * 0.05
    + response.usage.output_tokens / 1000000 * 0.40
)
print(f"{total_bill:.4f}")
print(f"{total_bill * 1470:.1f}")
```
