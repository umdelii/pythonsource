```python
from dotenv import load_dotenv, find_dotenv
from openai import OpenAI
from youtube_transcript_api import YouTubeTranscriptApi

load_dotenv(find_dotenv())
client = OpenAI()
model = "gpt-5-nano"
```

```python
video_id = "https://www.youtube.com/watch?v=008r2I5duAg"
ytt_api = YouTubeTranscriptApi()

# 자막 가져오기
transcript = ytt_api.fetch(video_id, languages=["en"])

# 자막을 텍스트로 정리
text = " ".join([t.text for t in transcript])

# 영어로 추출된 내용을 한국어로 번역 후 내용 요약 시키기
response = client.responses.create(
    model=model,
    input=text,
    instructions="""
    너는 영어 자막을 한국어로 번역하고, 핵심 내용을 요약할 수 있는 전문가이다.
    자막이 영어인 경우 실제 현대 네이티브 한국인들이 사용하는 한국어로 자연스럽게 번역하고, 주요 내용을 10줄 이내로 요약하라.
    요약은 가독성이 좋게 특정 형식(너가 생각했을때 좋은 방법으로)으로 작성하라.
    """,
)

print(response.output_text)
```
